```python
import seaborn as sns
import pandas as pd
import numpy as np
from transformers import pipeline
```

```python
df = pd.read_csv('/kaggle/input/intent-dataset/intent_dataset_extended_diverse.csv')
```

```python
confidence_threshold = 0.6

classifier = pipeline("zero-shot-classification")

def classify_intent(text):
    result = classifier(text, candidate_labels=['Greet', 'Farewell', 'Inquiry'], multi_label=False)
    intent = result['labels'][0]
    confidence = result['scores'][0]
    
    if confidence >= confidence_threshold:
        return intent, confidence
    else:
        return "NLU fallback: Intent could not be confidently determined", confidence
```

```python
# Test the model and classification function
test_texts = [
        "Hey, how's it going?",
        "What's the weather like tomorrow?",
        "Goodbye for now",
        "I need to know the time",
        "Where can I grab lunch?",
        "Hey, long time no see!",
        "Tell me a joke!",
        "Thanks, see you later",
        "Can you recommend a good book?",
        "How's your day been?"
    ]

for text in test_texts:
    intent, confidence = classify_intent(text)
    print(f"Text: '{text}' --> Intent: '{intent}' (Confidence: {confidence:.3f})")
```

```python
user_input = input("Enter your text: ")
intent, confidence = classify_intent(user_input)
print(f"Intent: '{intent}' (Confidence: {confidence:.3f})")
```
